Using RuTranscript

```python
from RuTranscript.src.ru_transcript import RuTranscript
```

```python
text = 'Как получить транскрипцию?'
accented_text_if_have = 'Как получи+ть транскри+пцию?'

ru_transcript = RuTranscript(text, accented_text_if_have)
ru_transcript.transcribe()

print(f'Full transcription: {ru_transcript.transcription}\nAllophones: {ru_transcript.allophones}')
```
